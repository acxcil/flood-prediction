```python
import matplotlib.pyplot as plt
import networkx as nx

# Build directed graph
G = nx.DiGraph()
nodes = [
    "Ingest Data",
    "Preprocess\n& Align",
    "Feature\nEngineering",
    "Fuzzy\nInference",
    "Stacked\nModel",
    "Expose via\nAPI"
]
edges = [
    ("Ingest Data", "Preprocess\n& Align"),
    ("Preprocess\n& Align", "Feature\nEngineering"),
    ("Feature\nEngineering", "Fuzzy\nInference"),
    ("Fuzzy\nInference", "Stacked\nModel"),
    ("Stacked\nModel", "Expose via\nAPI")
]
G.add_nodes_from(nodes)
G.add_edges_from(edges)

# Fixed layout coordinates
pos = {
    "Ingest Data": (0, 1),
    "Preprocess\n& Align": (2, 1),
    "Feature\nEngineering": (4, 1),
    "Fuzzy\nInference": (4, -1),
    "Stacked\nModel": (6, -1),
    "Expose via\nAPI": (8, -1)
}

# Draw oversized figure to avoid cropping
plt.figure(figsize=(12, 5))

# Draw nodes
nx.draw_networkx_nodes(
    G, pos,
    node_size=2500,
    node_color="#bbdefb",
    edgecolors="#1976d2"
)

# Draw labels
nx.draw_networkx_labels(G, pos, font_size=10, font_family="sans-serif")

# Draw directed edges
nx.draw_networkx_edges(
    G, pos,
    arrowstyle="-|>",
    arrowsize=20,
    edge_color="#1976d2",
    width=2
)

# Title and layout
plt.title("System Workflow Diagram", fontsize=16)
plt.axis("off")

# Expand margins to ensure nothing is cut off
plt.gca().margins(x=0.3, y=0.3)
plt.tight_layout()

# Save to file
# plt.savefig("figure_3_3_full.png", dpi=150)
plt.show()
```
